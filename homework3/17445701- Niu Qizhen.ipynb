```python
all_comments = []
for comments in df['comments']:
    all_comments.extend(comments.split())
```

```python
all_comments
```

```python
word_count = {}
```

```python
for word in all_comments:
    if word not in word_count:
        word_count[word] = 1
    else:
        word_count[word] = word_count[word]+1
```

```python
word_count
```

```python
pandas.Series(word_count).to_frame().reset_index()
```

```python
pandas.Series(word_count).sort_values(ascending=False)
```

```python
stop_words = set(['to', 'of', 'the', 'at', 'x', 'be'])
```

```python
stop_words
```

```python
df_word_counts = pandas.Series(word_count).to_frame().reset_index()
```

```python
df_word_counts.sort_values(by=0, ascending = False)
```

```python
def is_stop_word(x):
    return x in stop_words
df_word_counts[
    df_word_counts['index'].apply(is_stop_word)
].sort_values(by=0, ascending= False)
```
